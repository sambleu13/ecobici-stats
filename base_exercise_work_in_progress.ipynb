````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
#model libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble        import RandomForestClassifier
from sklearn.metrics         import roc_auc_score, roc_curve, mean_squared_error,mean_absolute_error
#To install sklearn, type in a terminal (anaconda prompt) the following line: conda install -c anaconda scikit-learn
import shap
#To install shap, type in a terminal (anaconda prompt) the following line: conda install -c conda-forge shap
````

Loading datasets

````python
#sam y melissa

rawdata = pd.read_csv(path)

print(rawdata.info())
print(rawdata.describe())

````

````python
rawdata= rawdata.drop(columns=['ID'])
````

````python
display(rawdata)
````

````python
#Outliers

#boxplots with names and thresholds values per column
def remove_outliers(column,name,threshold,df):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {name}')
    plt.show()
    print(df[column].max(), df[column].min())
    mean = np.mean(df[column])
    std = np.std(df[column])

    outliers = []
    for x in df[column]:
      z_score = (x - mean) / std
      if abs(z_score) > threshold:
        outliers.append(x)

    filtered_df = df[~df[column].isin(outliers)]
    print(threshold)
    print("Mean: ", mean)
    print("\nStandard deviation: ", std)
    print("\nOutliers  : ", outliers)
    #print("\nFiltered Data:\n", filtered_df)
    #print(filtered_df.describe())

    sns.boxplot(filtered_df[column])
    plt.title(f'Box Plot without Outliers of {name}')
    plt.show()
    print( filtered_df[column].max(), filtered_df[column].min())
    return filtered_df

selvars_with_outliersarray = [
    ["X1",'ROA',2],
    ["X6",'RE/TA',7],
    ["X7",'Operating Return on Assets',2],
    ["X8",'Equity Ratio',0.4],
    ["X18",'GPA',10],
    ["X19",'Gross Profit Margin',4],
    ["X25",'Equity-to-Asset Ratio',7],
    ["X48",'EBITDA/TA',2]]

for row in selvars_with_outliersarray:
    col1, col2, col3 = row
    no_outliers = remove_outliers(col1,col2,col3,rawdata)


no_outliers.describe()
````

````python
selvars = ['X1', 'X2', 'X6', 'X7', 'X8','X18', 'X19','X21', 'X25', 'X48']
selvars_name = ['X1-ROA', 'X2-Debt Ratio', 'X6-RE/TA', 'X7-Operating Return on Assets', 'X8-Equity Ratio','X18-GPA', 'X19-Gross Profit Margin','X21-Sales Growth', 'X25-Equity-to-Asset Ratio', 'X48-EBITDA/TA']

print(selvars)
print(selvars_name)
````

````python
#for to plot histograms of each chosen variable


for i,j in zip(selvars,selvars_name):
  plt.figure()
  plt.hist(no_outliers[i],bins=20,color="skyblue")
  plt.title('Distribution of ' + j)
  plt.xlabel(j)
  plt.ylabel('Values')
plt.show
````

````python
# Assuming no_outliers is your DataFrame and selvars is the list of selected variables
# no_outliers = pd.DataFrame({...})
# selvars = ['X1', 'X2', 'X6', 'X7', 'X8', 'X18', 'X19', 'X21', 'X25', 'X48']

# Plot all histograms in one merged histogram
plt.figure(figsize=(10, 8))
for i,j in zip(selvars,selvars_name):
    no_outliers[i].hist(bins=1, alpha=0.5, label=j)  # Use alpha for transparency

plt.title('Merged Histogram of Selected Variables')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
````

````python
correlation_fulldata = no_outliers.corr()

plt.figure(figsize=(10, 8))
axis_corr = sns.heatmap(
correlation_fulldata,
vmin=-1, vmax=1, center=0,
cmap=sns.color_palette("Blues", as_cmap=True),
square=True
)

plt.show()

print(no_outliers.info())
print(no_outliers.describe())
````

````python
#Heatmap

correlation_full_seldata = no_outliers[selvars].corr()
custom_palette = sns.color_palette(["#D8BFD8", "#87CEEB"])
plt.figure(figsize=(10, 8))
axis_corr = sns.heatmap(
correlation_full_seldata,
vmin=-1, vmax=1, center=0,
cmap=sns.blend_palette(custom_palette, as_cmap=True),
square=True, annot=True
)

plt.show()
````

````python
#filling NA's with the mean of the column:
df_clean= no_outliers.fillna(no_outliers.mean())
````

````python
#Preparing the data to feed the algorithm
#The dummy model only uses the first 10 features
print(selvars)

#Name of the target variable
target_name='TARGET'

# Converting the dataframe into numpy arrays
X = np.array(df_clean[selvars])
y = np.array(df_clean[target_name])

#Creating a train test subsets for the model
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
#Now the data is prepared to feed the algorithm

print(X_test)
print(X_test.shape)
print(X_train.shape)
print(X.shape)
print(y.shape)

````

````python
#Model definition
model_melissa = RandomForestClassifier(n_estimators=250, max_depth=15, criterion='gini')
````

````python
#Model Training
model_melissa.fit(X_train, y_train)
````

````python
#Training Metrics
score = model_melissa.score(X_train, y_train)
print("R-squared:", score)
````

````python
#prediction an train Metrics
y_train_pred = model_melissa.predict_proba(X_train)[:, 1]
print(y_train_pred)

AUC_train    = roc_auc_score(y_train, y_train_pred)
print("AUC:", AUC_train)
````

````python
#prediction an tests Metrics
y_test_pred = model_melissa.predict_proba(X_test)[:, 1]
print(y_train_pred)
AUC_test    = roc_auc_score(y_test, y_test_pred)
print("AUC:", AUC_test)
````

````python
#Prediction an test Metrics
y_pred = model_melissa.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("MSE: ", mse)
print("RMSE: ", mse*(1/2.0))
msa = mean_absolute_error(y_test, y_pred)
print("MSA: ", msa)
````

````python
# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)

# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
````

````python
explainer = shap.TreeExplainer(model_melissa)
#shap_values = explainer.shap_values(X_test)

shap_values = np.array(explainer.shap_values(X_test))
print(shap_values.shape)


#To plot feature importance as the horizontal bar plot we need to use summary_plot method:

shap.summary_plot(shap_values[:,:,1], X_test, np.array(selvars))

shap.summary_plot(shap_values[:,:,1], X_test, np.array(selvars),plot_type="bar")


````
